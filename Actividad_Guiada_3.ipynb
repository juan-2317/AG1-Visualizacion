```python
#Importamos modulos

import pandas as pd
import io
import requests
import seaborn as sns # Seaborn encapsula a mathplotlib -> Facilita la representación gráfica de los datos -> Acepta las estructuras de datos de pandas
import timeit
import matplotlib.pyplot as plt
```

```python
#Cargamos el data set de los pasajeros del Titanic

url="https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
s=requests.get(url).content
titanic=pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
titanic.info()
```

```python
titanic.head(5)
```

```python
titanic.rename(columns={'fare': 'tarifa'}, inplace=True)
```

```python
titanic.rename(columns={'class': 'clase'}, inplace=True)
```

```python
def is_old_func(row):
  return row['age'] > 60

titanic['is_old'] = titanic.apply(is_old_func, axis='columns')
```

```python
titanic.eval ( ' is_baby = age< 15 ' , inplace = True)
```

```python
#Consulta con condiciones

titanic[

  (titanic.sex == 'female')

  & (titanic['clase'].isin(['First', 'Third']))

  & (titanic.age > 45 )

  & (titanic.survived == 0)

]
```

```python
#Distribución de las clases

sns.countplot(x="pclass", data=titanic)
```

```python
#Distribución de la edad(ege)
sns.distplot(titanic.age.dropna( ))
plt.show( )
```

```python
#BoxPlot de la edad por sexo y clase

with sns.axes_style(style='ticks'):

  ax = sns.catplot("sex", "age", "clase", data=titanic, kind="box")
```

```python
#Distribución cruzada de Edad y Tarifa
sns.jointplot(x='tarifa',y='age',data=titanic) # Diagrama de dispersión
```

```python
sns.jointplot(x='tarifa',y='age',data=titanic[(titanic.sex == 'female')]) # Filtro de tarifa y edad solamente de mujeres
```

```python
sns.jointplot(x='tarifa',y='age',data=titanic[(titanic.sex == 'male')]) # Filtro de tarifa y edad solamente de hombres
```

```python
#Cambiamos el font
sns.set(font_scale=1)
#FacetGrid - Construir una matriz de gráficos
g = sns.FacetGrid(titanic, row='survived',col='clase')
g.map(sns.distplot, "age") # Colocamos en esta función la variable x, para el ejemplo es la variable "age"
plt.show()
```

```python
#Diagrama de dispersion con Distribucion de cada variable: fare(precio)/age(edad)
sns.jointplot(data=titanic, x='age', y='tarifa', kind='reg', color='b')
plt.show()
```
